```python
import pandas as pd
perlogs                 = pd.read_csv("C:\TANIGUCHIJUN_20170217-000001\Performance_Counter.csv")
AvailableMBytes         = perlogs["TANIGUCHIJUNMemoryAvailableMBytes"]
DiskTime                = perlogs["TANIGUCHIJUNPhysicalDisk(0C:)%DiskTime"]
Pages_sec               = perlogs["TANIGUCHIJUNMemoryPages/sec"]
BytesTotal_sec          = perlogs["TANIGUCHIJUNNetworkInterface(TeredoTunnelingPseudo-Interface)BytesTotal/sec"]
Avg_DiskQueueLength     = perlogs["TANIGUCHIJUNPhysicalDisk(_Total)Avg.DiskQueueLength"]
CurrentDiskQueueLength  = perlogs["TANIGUCHIJUNPhysicalDisk(_Total)CurrentDiskQueueLength"]
DiskWrites_sec          = perlogs["TANIGUCHIJUNPhysicalDisk(_Total)DiskWrites/sec"]
Processor_ProcessorTime = perlogs["TANIGUCHIJUNProcessor(_Total)%ProcessorTime"]
Process_ProcessorTime   = perlogs["TANIGUCHIJUNProcess(McTkSchedulerService)%ProcessorTime"]
UserTime                = perlogs["TANIGUCHIJUNProcessor(_Total)%UserTime"]
InterruptTime           = perlogs["TANIGUCHIJUNProcessor(_Total)%InterruptTime"]
```
